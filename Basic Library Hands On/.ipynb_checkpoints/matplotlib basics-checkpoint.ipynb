## First time with Matplotlib


```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,9)
```

```python
x
```

```python
y = x**3
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("this is x")
plt.ylabel("y")
```

```python
plt.plot([1,2,3,4], [5,6,7,8], label = "population")
plt.plot([1,2,3,4], [2,4,6,8], label = "consumption")
plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("Two lines")
plt.legend()
```

```python
plt.bar([1,2,3,4], [5,6,7,8], label = "population", width=0.5)

plt.xlabel("this is x")
plt.ylabel("this is y")
plt.title("Two lines")
```

```python
myDays = [1,2,3,4,5]
organic = [8, 9, 8, 3, 3]
ads = [8, 9, 7, 12, 8]
affiliate = [3, 5, 5, 4, 3]
seller = [9, 5, 8, 9, 14]
```

```python
plt.plot([], [], color='#F97F51', label='organic')
plt.plot([], [], color='#1B9CFC', label='ads')
plt.plot([], [], color='#F8EFBA', label='affiliate')
plt.plot([], [], color='#58B19F', label='seller')

plt.stackplot(myDays, organic,ads,affiliate,seller, colors=['#F97F51', '#1B9CFC', '#F8EFBA', '#58B19F'])
plt.xlabel('Number of days')
plt.ylabel('Sales')
plt.legend()
```

```python
mypieLabels = ['FB', 'GOOGLE', 'LinkedIn', 'Pinterest', 'Twitter']
```

```python
piecolors = ['#F97F51', '#1B9CFC', '#FD7272', '#58B19F', '#55E6C1']
```

```python
plt.pie(ads, labels=mypieLabels, colors=piecolors, explode=(0, 0, 0, 0, 0),shadow = True,labeldistance=1.1,radius=2)
```
